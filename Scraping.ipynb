Sebelum melakukan analisis sentimen kita akan melakukan proses scraping. Proses ini ini bertujuan untuk mengumpulkan data ulasan pengguna aplikasi tokopedia dari google play store menggunakan library google-play-scraper. Fungsi reviews() digunakan untuk mengambil hingga 10.000 review dengan pengaturan bahasa bahasa indonesia (lang='id') dan lokasi negara indonesia (country='id'). Data yang diperoleh disimpan ke dalam dataframe dengan kolom seperti userName, score, dan content (isi ulasan). Kemudian, data hasil scraping tersebut disimpan dalam file CSV bernama tokopedia_reviews.csv agar dapat digunakan untuk proses preprocessing dan pelabelan sentimen.

```python
from google_play_scraper import reviews
import pandas as pd
```

```python
result, _ = reviews(
    'com.tokopedia.tkpd',
    lang='id',
    country='id',
    count=10000,
)

data = pd.DataFrame(result)
data[['userName', 'score', 'content']].to_csv("tokopedia_reviews.csv", index=False)
```
